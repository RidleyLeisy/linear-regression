```python
import scipy.stats as stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing/Cleaning Automobile Dataset

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',header=None)
```

```python
# cleaning dataset and pulling out missing price information into a dataframe called "missing_values"

cols = ['price','horsepower']
missing_values = df.loc[df[25] == '?']
df = df[[25,21]].replace('?',np.nan).dropna()
df.columns = cols
missing_values = missing_values[[25,21]].replace('?',np.nan)
missing_values.columns = cols

# changing datatypes to numeric
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['price'] = pd.to_numeric(df['price'])
missing_values['horsepower'] = pd.to_numeric(missing_values['horsepower'])
```
